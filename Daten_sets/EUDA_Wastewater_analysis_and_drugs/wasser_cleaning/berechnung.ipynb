```python
import pandas as pd

staedte=['Chemnitz','Dresden','Leipzig','Berlin','Hamburg','München','Köln','Frankfurt','Stuttgart','Düsseldorf']
```

```python
for stadt in staedte:
    df = pd.read_csv(f'{stadt}.csv', sep=';')
    df['Veränderung (%)'] = df['Daily mean'].pct_change() * 100
    

# 3. Ergebnis anzeigen
print(df)
```
