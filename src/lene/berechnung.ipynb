```python
import pandas as pd

staedte=['Dresden','Chemnitz','Dortmund','Erfurt','Magdeburg','Munich','Nuremberg','Saarbrücken']
```

```python
all_changes = []

for stadt in staedte:
    df = pd.read_csv(f'../../Daten_sets/wasser_cleaning/{stadt}/{stadt}.csv', sep=';')
    df['Change (%)'] = df['Daily mean'].pct_change() * 100
    df_changes = df.pivot(index='City', columns='Year', values='Change (%)')
    all_changes.append(df_changes)

df_final = pd.concat(all_changes)
df_final = df_final.round(2)

print(df_final)

df_final.to_csv('../../Daten_sets/wasser_cleaning/prozentuale_aenderung.csv', sep=';')
```
