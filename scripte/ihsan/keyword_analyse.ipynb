```python
import pandas as pd
import re
```

```python
def Keyword_analyse(csv):
    
    #savong the path for later regex expression
    pfad = csv

    #reading the csv
    csv = pd.read_csv(csv)

    #converting to the Data Drame
    csv = pd.DataFrame(csv)
    
    #checking if there are empty or non string values 
    csv["title"] = csv["title"].fillna("").astype(str)
    csv["abstract"] = csv["abstract"].fillna("").astype(str)
    csv["datum"] = csv["datum"].fillna("").astype(str)

    # Defining the Key Words
    key_words = ["Drogen", "Droge", "kokain"]
    
    #Defining the regex rules
    rule_date = r"(\d{2}\.\d{2}\.\d{4})"  
    rule_time = r"(\d{2}:\d{2})"

    # Redifining the sub parts of "datum" into Date and Time
    csv["Datum"] = csv["datum"].str.extract(rule_date)
    csv["time"] = csv["datum"].str.extract(rule_time)

    # Old "datum" colum is dalated
    csv = csv.drop(columns="datum")
    
    # Adding essential colums
    csv["Text Length"] = 0
    csv["Keyword Count"] = 0

    # Normalised 
    normalised_key_words = [i.lower() for i in key_words]





    # Loop for counting the length and keywords 
    for row in range(len(csv)):

        word_bag = csv.loc[row ,"title"]+ csv.loc[row ,"abstract"]

        word_bag_list = word_bag.split(" ")

        clean_list = [w.strip('.,!?:;"').lower() for w in word_bag_list]

        final_list = [w for w in clean_list if w.isalpha()] # clearing the numbers


        
        normalised_word_bag_list = [word.lower() for word in final_list] # normilse

        csv.loc[row ,"Text Length"] = len(csv.loc[row,"title"]+csv.loc[row,"abstract"]) # detecting the text length 

        for word in normalised_word_bag_list:
            if word in normalised_key_words:
                csv.loc[row, "Keyword Count"] += 1  
    
    #creating the file name
    Daten_set_name = pfad.split('/')[-1]
    output_path = f"/tmp/{Daten_set_name}" # ! Attention: Saving to the tmp folder !
    
    return csv.to_csv(f"{output_path}", index=False)

```

```python
folders_csv_in_repo = ["Chemnitz_blaulicht_scrape_2025-11-23.csv","Chemnitz_blaulicht_scrape_2025-12-07.csv","Dortmund_blaulicht_scrape_2025-11-23.csv","Dortmund_blaulicht_scrape_2025-12-07.csv","Duelmen_blaulicht_scrape_2025-12-07.csv","Duelmen_blaulicht_scrape_2025-12-09.csv","Erfurt_blaulicht_scrape_2025-12-09.csv","Nuernberg_blaulicht_scrape_2025-12-09.csv","Saarbruecken_blaulicht_scrape_2025-12-09.csv","Stuttgart_blaulicht_scrape_2025-12-09.csv"]

for file in folders_csv_in_repo:
    pfad = f"https://raw.githubusercontent.com/ichzann/DIS08_Data_modeling_7er_Gruppe/main/Daten_sets/blaulicht_scraping/{file}"
    Keyword_analyse(pfad)
```
